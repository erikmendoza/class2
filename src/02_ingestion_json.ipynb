```python
import pandas as pd 
import requests 

url = "https://rickandmortyapi.com/api/character"

response = requests.get(url)
data = response.json()
data = data["results"]

df = pd.json_normalize(data)

#display(df.head(6))

#df.loc[4,'episode']


df['apariciones'] = df['episode'].apply(len)

df['ha'] = df['species'].str[0]

df['boss'] = df['name'].apply(lambda x: "jefe" if "Rick" in x else "NO")

df['boss'] = df['boss'].apply(str.upper)


df['is_alien'] = df['species'].apply(lambda x: "YES" if "Alien" in x else "NO")

#df['profile'] = df[df['name','species']]

#df['profile'] = df['name'],df['species']











#df['profile'] = df['name'].apply(str.upper) +" (" + df['species'] + ")"


df['profile'] = df.apply(lambda x: f"{x['name']} | {x['species']}", axis=1)






df_danger = df[df['status'] == "Alive"]

df_danger = df_danger[df_danger['apariciones'] > 30 ]








df.to_csv("../data/reporte_rick.csv")
df_danger.to_csv("../data/danger_humans.csv")






df['is_morty'] = df['name'].apply(lambda x: x.split(" "))
df['is_morty'] = df['is_morty'].apply(lambda x: "is Morty" if "Morty" in x else "No")




#df['category'] = df['apariciones'].apply(lamda x: "")


#reporte = df.groupby('origin.name')['status'].apply(lambda x: x == 'Dead')

#reporte = df.groupby('origin.name')['status'].count("Dead")

#reporte = df.groupby('origin.name')['status'].apply(lambda x: (x == 'Dead').count())


df['dead'] = df['status'].apply(lambda x: 1 if 'Dead' in x else 0)

report = df.groupby('origin.name')['dead'].agg(['count','sum'])

report['percentage'] = report['sum'] / report['count'] * 100

report.columns = ['Total_characters', 'Dead', 'Percentage']

display(report)


#display(df['origin.name'] + df['status'])

#display(df_danger.head(20))
```
